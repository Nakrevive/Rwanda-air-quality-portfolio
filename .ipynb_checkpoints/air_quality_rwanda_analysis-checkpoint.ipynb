```python
import pandas as pd

df = pd.read_csv(r"C:\Users\educa\Desktop\air_quality_africa_project\data\Rwanda_air_quality.csv")
print(df.head())
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['country_iso'] = df['country_iso'].fillna("RW")
df['isMobile'] = df['isMobile'].fillna(False)
df['isMonitor'] = df['isMonitor'].fillna(False)


```

```python
df.isna().sum()
```

```python
df['parameter'].unique()
```

```python
df['datetimeLocal'] = pd.to_datetime(df['datetimeLocal'])
df['datetimeUtc'] = pd.to_datetime(df['datetimeUtc'])
```

```python
df['date'] = df['datetimeLocal'].dt.date
df['year'] = df['datetimeLocal'].dt.year
df['month'] = df['datetimeLocal'].dt.month
df['day'] = df['datetimeLocal'].dt.day
df['hour'] = df['datetimeLocal'].dt.hour
```

```python
df['value'].describe()
```

```python
daily_avg = df.groupby('date')['value'].mean()
daily_avg
```

```python
import matplotlib.pyplot as plt

daily_avg.plot(kind='line', marker='o', figsize=(8,4))
plt.title('Daily Average Value')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(daily_avg.index, daily_avg.values, marker='o')
ax.set_title('Daily Average Values')
ax.set_xlabel('Date')
ax.set_ylabel('Value')
ax.grid(True)

# Format dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.show()
```

```python
df['datetimeLocal'] = pd.to_datetime(df['datetimeLocal'])
df['hour'] = df['datetimeLocal'].dt.hour

hourly_avg = df.groupby('hour')['value'].mean()

hourly_avg.plot(kind='line', marker='o', figsize=(8,4))
plt.title('Average PM2.5 by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('PM2.5 Value (µg/m³)')
plt.grid(True)
plt.show()
```

```python
station_avg = df.groupby(['date', 'location_name'])['value'].mean().unstack()
station_avg.plot(figsize=(10,5))
plt.title('Daily PM2.5 Averages by Station')
plt.xlabel('Date')
plt.ylabel('PM2.5 Value (µg/m³)')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px

fig = px.line(
    x=daily_avg.index,
    y=daily_avg.values,
    labels={'x':'Date', 'y':'PM2.5 Value (µg/m³)'},
    title='Daily Average PM2.5 Values',
    markers=True
)
fig.show()
```
